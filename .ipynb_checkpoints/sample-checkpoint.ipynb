```python
import streamlit as st
import pandas as pd
import plotly.express as px

# Load Data
df = pd.read_excel("data/Test Data 22 Intern (1).xlsx")

# Data Cleaning
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()

# Sidebar Filters
st.sidebar.header("Filters")

year_filter = st.sidebar.multiselect(
    "Select Year",
    options=df['Year'].unique(),
    default=df['Year'].unique()
)

zone_filter = st.sidebar.multiselect(
    "Select Zone",
    options=df['Zone'].unique(),
    default=df['Zone'].unique()
)

state_filter = st.sidebar.multiselect(
    "Select State",
    options=df['State'].unique(),
    default=df['State'].unique()
)

# Apply Filters
filtered_df = df[
    (df['Year'].isin(year_filter)) &
    (df['Zone'].isin(zone_filter)) &
    (df['State'].isin(state_filter))
]

# KPIs
total_sales = filtered_df['Sales value'].sum()
pending_bills = filtered_df[filtered_df['Sales value'] < 0]['Sales value'].sum()

st.title("📊 Sales Dashboard")

col1, col2 = st.columns(2)
col1.metric("Total Sales", f"{total_sales:,.0f}")
col2.metric("Pending Bills", f"{pending_bills:,.0f}")

# ---- Yearly Sales ----
yearly = filtered_df.groupby('Year')['Sales value'].sum().reset_index()
fig_year = px.bar(yearly, x='Year', y='Sales value', title="Yearly Sales")
st.plotly_chart(fig_year)

# ---- Quarterly Sales ----
filtered_df['Quarter'] = filtered_df['Date'].dt.to_period('Q').astype(str)
quarterly = filtered_df.groupby('Quarter')['Sales value'].sum().reset_index()

fig_quarter = px.pie(quarterly, names='Quarter', values='Sales value', title="Quarterly Sales")
st.plotly_chart(fig_quarter)

# ---- Zonal Sales ----
zone_sales = filtered_df.groupby('Zone')['Sales value'].sum().reset_index()

fig_zone = px.bar(zone_sales, x='Zone', y='Sales value', title="Zonal Sales", color='Zone')
st.plotly_chart(fig_zone)

# ---- Sales by Representative ----
rep_sales = filtered_df.groupby('External Sales Representative')['Sales value'].sum().reset_index()

fig_rep = px.bar(rep_sales, x='External Sales Representative', y='Sales value', title="Sales by Representative")
st.plotly_chart(fig_rep)

# ---- Product Sales ----
product_sales = filtered_df.groupby('Product Description')['Sales value'].sum().reset_index()

fig_product = px.bar(product_sales, x='Product Description', y='Sales value', title="Product Sales")
st.plotly_chart(fig_product)
```
